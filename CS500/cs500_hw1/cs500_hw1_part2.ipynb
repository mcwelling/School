```python
%%sql sqlite:///routes.db
--
-- PROBLEM 1. [5 pts]
--   Find cities that have routes that depart from that city to at least five unique destinations

select
dep.city as City
,dep.Country
,count(distinct arr.name) as Unique_Destinations
from airports dep
join routes r on dep.airportID = r.sourceID
join airports arr on r.destinationID = arr.airportID
group by dep.city, dep.Country
having count(distinct arr.name) >= 5;
```

```python
%%sql sqlite:///routes.db
--
-- PROBLEM 2. Cities with no return flights. [5 pts]
--   Retrieve routes between two cities that do not have a return route

select distinct
c1.city as departure_city
,c2.city as arrival_city
from routes r1
left join routes r2 on r1.sourceID = r2.destinationID and r1.destinationID = r2.sourceID
left join airports c1 on r1.sourceID = c1.airportID
left join airports c2 on r1.destinationID = c2.airportID
where r2.destinationID is null
order by c1.city;
```

```python
%%sql sqlite:///routes.db
--
-- PROBLEM 3. Cities with round-trip by two airlines. [10 pts]
--   Find cities that only have round-trip routes operated by different airlines.
--   In other words, we want to find cities that that have round-trip flights between them, but 
--   there are no round-trip flights where both routes are operated by a single airline

select distinct
c1.city as departure_city
,c2.city as arrival_city
from routes r1
join routes r2 on r1.sourceID = r2.destinationID and r1.destinationID = r2.sourceID
join airports c1 on r1.sourceID = c1.airportID
join airports c2 on r1.destinationID = c2.airportID
where r1.airlineID <> r2.airlineID
order by c1.city
;
```

```python
%%sql sqlite:///routes.db
--
-- PROBLEM 4. At least two options for a single-airline round-trip. [10 pts]
--   Find cities where there are at least two companies, where 
--   each company operates round-trip routes between two cities
select
departure_city
,arrival_city
,count(distinct dep_airline1) Companies
from (
select distinct
c1.city as departure_city
,c2.city as arrival_city
,r1.airlineID dep_airline1
,r2.airlineID ret_airline1
,r1.airlineID dep_airline2
,r2.airlineID ret_airline2
from routes r1
join routes r2 on r1.sourceID = r2.destinationID and r1.destinationID = r2.sourceID
join airports c1 on r1.sourceID = c1.airportID
join airports c2 on r1.destinationID = c2.airportID
where r1.airlineID = r2.airlineID)
group by departure_city, arrival_city
having count(distinct dep_airline1) > 1;
```

```python
%%sql sqlite:///chinook.db
--
-- PROBLEM 5. Highest-grossing album. [5 pts]
--   Find five albums that produced the largest amount in total sales

select
a.albumID
, a.title
, sum((ii.unitPrice * ii.quantity)) as Total
from invoice_items ii
join tracks t on ii.trackId = t.trackId
join albums a on t.albumId = a.albumID
group by a.albumID, a.title
order by Total desc
limit 5
;
```

```python
%%sql sqlite:///chinook.db
    
-- PROBLEM 6. One-hit wonders. [10 pts]
--   Retrieve all artists for which the following condition is true:
--      Let "at1" denote the total amount in sales for the highest-grossing track of an artist
--      Let "at2" denote the total amount in sales for the second highest-grossing track of the artist
--      Select artists when at1>at2*2 
--         (i.e., sales for the artist''s second best track were less than 50% of the top track)


select
artist_name
,trackID
,track_name
,max(total)
,min(total)
from
(
select artist_name, trackID, track_name, Total
    from (
        select 
            tsales.artist_name
            ,tsales.trackID
            ,tsales.track_name
            ,tsales.Total 
              ,row_number() over(
                 partition by artist_name
                 order by Total desc
                ) RowNum
        from 
            (select
            art.name as artist_name
            ,t.trackID
            ,t.name as track_name
            ,sum((ii.unitPrice * ii.quantity)) as Total
            from invoice_items ii
            join tracks t on ii.trackID = t.trackID
            left join albums a on t.albumID = a.albumID
            left join artists art on a.artistID = art.artistID
            group by art.name, t.trackID, t.name) tsales
        
        ) rs where RowNum <= 2
)
group by artist_name, trackID, track_name
having max(total) > (min(total) * 2);

--it appears there are none
```

```python
%%sql sqlite:///chinook.db
--
-- PROBLEM 7. Custumers with shared music preference. [10 pts]
--   Identify 5 customer pairs that share the most number of tracks in their purchase history


select
composite as customerID_pairs
,count(*)
from (
select
ii.trackid
,max(i.customerid)||', '||min(i.customerid) as composite
from invoices i
join invoice_items ii on i.invoiceid = ii.invoiceid
group by ii.trackid
having count(*) > 1)
group by composite
order by count(*) desc
limit 5;
```
